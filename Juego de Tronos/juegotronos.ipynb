```python
# IMPORT KML DRIVER

gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

# READ KML file to a geopandas dataframe 

geo_df = gpd.read_file('Game of Thrones filming locations in Iceland.kml',driver='KML')

# Create Pandas Dataframe from GeoPandas 

df_iceland= pd.DataFrame(geo_df)

# Extract latitude and longitude from the KML geometry column
df_iceland['lat'] = df_iceland.geometry.apply(lambda p: p.y)
df_iceland['lon'] = df_iceland.geometry.apply(lambda p: p.x)
df_iceland["country"] = "Iceland"
```

```python
df_iceland
```

```python
geo_df2 = gpd.read_file('Game of Thrones filming locations in Malta.kml',driver='KML')

# Create Pandas Dataframe from GeoPandas 

df_malta= pd.DataFrame(geo_df2)

# Extract latitude and longitude from the KML geometry column
df_malta['lat'] = df_malta.geometry.apply(lambda p: p.y)
df_malta['lon'] = df_malta.geometry.apply(lambda p: p.x)
df_malta["country"] = "Malta"
```

```python
df_malta.head()
```

```python
geo_df3 = gpd.read_file('Game of Thrones filming locations Spain.kml',driver='KML')

# Create Pandas Dataframe from GeoPandas 

df_spain= pd.DataFrame(geo_df3)

# Extract latitude and longitude from the KML geometry column
df_spain['lat'] = df_spain.geometry.apply(lambda p: p.y)
df_spain['lon'] = df_spain.geometry.apply(lambda p: p.x)
df_spain["country"] = "Spain"
```

```python
geo_df4 = gpd.read_file('Game of Thrones locations in Northern Ireland.kml',driver='KML')

# Create Pandas Dataframe from GeoPandas 

df_ireland= pd.DataFrame(geo_df4)

# Extract latitude and longitude from the KML geometry column
df_ireland['lat'] = df_ireland.geometry.apply(lambda p: p.y)
df_ireland['lon'] = df_ireland.geometry.apply(lambda p: p.x)
df_ireland["country"] = "Ireland"
```

```python
df_mapa = pd.concat([df_malta, df_ireland, df_spain, df_iceland], axis= 0)
```

```python
df_mapa.drop("Description", inplace= True, axis= 1)
```
